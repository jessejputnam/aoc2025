```python
def get_end_idx(count: int, i: int):
    return -1 * (count - 1 - i) if i != (count - 1) else None

def get_largest_jolt(row: str, count: int):
    jolt_output = ['0'] * count
    idx = 0
    for i in range(count):
        idx_add = 0
        end = get_end_idx(count, i)
        for j, n in enumerate(row[idx:end]):
            if n > jolt_output[i]:
                jolt_output[i] = n
                idx_add = j
        idx += (idx_add + 1)
    return int(''.join(jolt_output))
```
