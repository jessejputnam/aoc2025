```python
type Coord = tuple[int,int]
```

```python
def set_vals(x: str) -> Coord:
    a, b = x.split(',')
    return (int(a), int(b))

def get_coords(data: list[str]) -> list[Coord]:
    return [set_vals(x) for x in data]

def build_board(coords: set[Coord]):
    return {
        "data": coords,
        "max_width": max([x[0] for x in coords]) + 3,
        "max_height": max([x[1] for x in coords]) + 2,
    }

def get_area(c1: Coord, c2: Coord):
    w = abs(c2[0] - c1[0]) + 1
    l = abs(c2[1] - c1[1]) + 1
    return w * l
```

```python
def part_one(data: list[str]):
    coords = get_coords(data)
    max_area = 0
    for i in range(len(coords)):
        for j in range(i + 1, len(coords)):
            max_area = max(get_area(coords[i], coords[j]), max_area)
    return max_area

part_one(data)
```
