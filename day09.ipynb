```python
type Coord = tuple[int,int]

class Boundary:
    def __init__(self, c1: Coord, c2: Coord, type: str):
        self.type = type
        self.idx = c1[1] if type == 'row' else c1[0]
        self.min = min(c1[0], c2[0]) if type == 'row' else min(c1[1], c2[1])
        self.max = max(c1[0], c2[0]) if type == 'row' else max(c1[1], c2[1])

    def __str__(self):
        return f"{self.type}: {{ idx: {self.idx}, min: {self.min}, max: {self.max} }}"
```

```python
def set_vals(x: str) -> Coord:
    a, b = x.split(',')
    return (int(a), int(b))

def get_coords(data: list[str]) -> list[Coord]:
    return [set_vals(x) for x in data]

def build_board(coords: list[Coord]):
    return {
        "coords": set(coords),
        "max_width": max([x[0] for x in coords]) + 3,
        "max_height": max([x[1] for x in coords]) + 2,
    }

def get_area(c1: Coord, c2: Coord):
    w = abs(c2[0] - c1[0]) + 1
    l = abs(c2[1] - c1[1]) + 1
    return w * l

def get_bounds(coords: list[Coord]):
    rows = []
    cols = []

    for i in range(len(coords)):
        for j in range(i + 1, len(coords)):
            c1, c2 = coords[i], coords[j]
            if c2[0] == c1[0]:
                boundary = Boundary(c1, c2, 'col')
                cols.append(boundary)
            if c2[1] == c1[1]:
                boundary = Boundary(c1, c2, 'row')
                rows.append(boundary)            
    return sorted(rows, key=lambda x: x.idx), sorted(cols, key=lambda x: x.idx)
```

```python
def part_one(data: list[str]):
    coords = get_coords(data)
    max_area = 0
    for i in range(len(coords)):
        for j in range(i + 1, len(coords)):
            max_area = max(get_area(coords[i], coords[j]), max_area)
    return max_area

part_one(data)
```
