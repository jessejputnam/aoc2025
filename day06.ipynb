```python
def part_one(data: list[str]):
    sets = [x.split() for x in data]
    result = 0
    for i in range(len(sets[0])):
        op = sets[-1][i]
        output = 0 if op == '+' else 1
        for j in range(len(sets) - 1):
            if op == '+': output += int(sets[j][i])
            else: output *= int(sets[j][i])
        result += output

    return result
    
part_one(test_data)
```

```python
class ColumnBuilder:
    def __init__(self, data: str):
        self.data = data
        self.cols = []
        self.row = []
        self.item = ''
        self.col_widths = self.get_col_widths()
        self.col_fill = self.col_widths[0]
        self.build_rows()
        self.ops = [x.replace('_', '') for x in self.cols[-1]]
        self.pivot()

    def get_col_widths(self) -> list[int]:
        ops = [x for x in self.data.split('\n') if x != ''][-1]
        col_widths = []
        cur_width = 0
        for c in ops[1:]:
            if c != ' ':
                col_widths.append(cur_width)
                cur_width = 0
            else:
                cur_width += 1
        col_widths.append(cur_width + 1)
        return col_widths
    
    def build_rows(self):
        counter = 0
        for c in self.data:
            if c == '\n':
                self.add_item()
                counter = 0
                self.col_fill = self.col_widths[counter]
                self.cols.append(self.row)
                self.row = []
            elif self.col_fill == 0:
                self.add_item()
                counter += 1
                self.col_fill = self.col_widths[counter]
            else:
                self.col_fill -= 1
                self.item = f"{self.item}{c.replace(" ", "_")}"
        if len(self.row) > 0:
            self.add_item()
            self.cols.append(self.row)

    def add_item(self):
        self.row.append(self.item)
        self.item = ''

    def pivot(self):
        new_cols = []
        for i in range(len(self.cols[0]) -1, -1, -1):
            col = []
            for j in range(len(self.cols)):
                col.append(self.cols[j][i])
            new_cols.append(col[:-1])
        self.cols = new_cols
```

```python
from lib import printf
def consume_column(col: list[int], op: str):
    output = 0 if op == '+' else 1
    for i in col:
        if op == '+': output += i 
        else: output *= i
    return output

def build_final_col(cols: list[list[str]], i: int):
    final_col = []
    for k in range(len(cols[i][0]) - 1, -1, -1):
        col_num = ''
        for j in range(len(cols[0])):
            col_num = f"{col_num}{(cols[i][j][k])}"
        final_col.append(int(col_num.replace("_", "")))
    return final_col

def solve_problem(c: ColumnBuilder, n, i):
    return consume_column(build_final_col(c.cols, i), c.ops[n - 1 - i])


def part_two(data: list[str]):
    raw_data = data[0]
    c = ColumnBuilder(raw_data)
    n = len(c.cols)
    solutions = [solve_problem(c, n, i) for i, x in enumerate(c.cols)]
    return sum(solutions)
  

# part_two(test_data)
part_two(data)
```
