```python
import math 

def is_invalid(i: int):
    s = str(i)
    return s[:int(len(s) / 2)] == s[int(len(s) / 2):]

def get_factors(num: int):
    factors = []
    for i in range(2, int(math.sqrt(num)) + 1):
        if num % i == 0:
            factors.append([i, num // i])
    return factors

def is_all_same(s: str):
    for c in s:
        if c != s[0]:
            return False
    return True    

def is_invalid_sub(factors: list[int], s: str):
    check = s[:factors[0]]
    for j in range(1, factors[1]):
        start = factors[0] * j
        end = start + factors[0]
        if s[start:end] != check:
            return False
    return True

def is_invalid_sub_reverse(factors: list[int], s: str):
    check = s[:factors[1]]
    for j in range(1, factors[0]):
        start = factors[1] * j
        end = start + factors[1]
        if s[start:end] != check:
            return False
    return True

def is_invalid_improved(i: int):
    s = str(i)
    factors = get_factors(len(s))
    if is_all_same(s):
        return True
    for f in factors:
        is_invalid = is_invalid_sub(f, s)
        is_invalid_reverse = is_invalid_sub_reverse(f, s)
        if is_invalid or is_invalid_reverse:
            return True
    return False
```
