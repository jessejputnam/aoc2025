```python
def is_invalid(manifold: list[str], row: int, col: int):
    return row < 0 or col < 0 or row >= len(manifold) or col >= len(manifold[0])

def tachyon_pass(manifold: list[str], row: int, col: int, split: set):
    if is_invalid(manifold, row, col):
        return
    
    while True:
        if is_invalid(manifold, row, col):
            return
        if manifold[row][col] == '^':
            if f"{row}|{col}" in split:
                return
            break
        row += 1
    split.add(f"{row}|{col}")
    tachyon_pass(manifold, row, col - 1, split)
    tachyon_pass(manifold, row, col + 1, split)

def quantum_pass(manifold: list[str], row: int, col: int, visits: dict[tuple[int, int], int]) -> int:
    if row >= len(manifold):
        return 1
    
    key = (row, col)

    if key in visits:
        return visits[key]
    
    if manifold[row][col] == '.':
        space_val = quantum_pass(manifold, row + 1, col, visits)
        visits[key] = space_val
        return space_val

    space_val = quantum_pass(manifold, row, col - 1, visits) + quantum_pass(manifold, row, col + 1, visits)
    visits[key] = space_val
    return space_val
```

```python
def part_one(data: list[str]):
    starting_index = data[0].index('S')
    split = set()
    tachyon_pass(data, 0, starting_index, split)
    return len(split)
part_one(data)
```

```python
def part_two(data: list[str]):
    starting_index = data[0].index('S')
    visits = {}
    return quantum_pass(data, 0, starting_index, visits)

print(part_two(data)) 
```
