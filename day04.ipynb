```python
def is_valid(i: int, j: int, x: int, y: int):
    return i >= 0 and j >= 0 and i < x and j < y

def is_paper(d: list[str], i: int, j: int, x: int, y:int):
    if not is_valid(i, j, x, y):
        return 0
    return 1 if d[i][j] == '@' else 0

def is_accessible(d: list[str], i: int, j: int, x: int, y:int):
    n = is_paper(d, i - 1, j, x, y)
    nw = is_paper(d, i - 1, j - 1, x, y)
    w = is_paper(d, i, j - 1, x, y)
    sw = is_paper(d, i + 1, j - 1, x, y)
    s = is_paper(d, i + 1, j, x, y)
    se = is_paper(d, i + 1, j + 1, x, y)
    e = is_paper(d, i, j + 1, x, y)
    ne = is_paper(d, i - 1, j + 1, x, y)
    return sum([n, nw, w, sw, s, se, e, ne]) < 4

def get_accessible_paper_locations(data: list[str]):
    x = len(data)
    y = len(data[0])
    loc = []

    for i in range(x):
        for j in range(y):
            if data[i][j] == '@':
                if is_accessible(data, i, j, x, y):
                    loc.append([i, j])
    return loc
```

```python
## PART 2
def full_removal(d: list[str]):
    removals = 0
    while True:
        locs = get_accessible_paper_locations(d)
        if len(locs) == 0:
            return removals
        removals += len(locs)
        for loc in locs:
            left = d[loc[0]][:loc[1]] 
            right = d[loc[0]][loc[1] + 1:]
            d[loc[0]] = f"{left}.{right}" 
full_removal(data)
        
```
