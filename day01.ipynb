```python
from lib import get_data

test_data = get_data(1, '\n', 'test')
data = get_data(1, '\n')

def rotate_dial(cmd: str, current_val: int) -> int:
    direction = 1 if cmd[0] == 'R' else -1
    rotation = int(cmd[1:]) * direction
    return rotation + current_val

def convert_rotation(dial: int) -> int:
    return dial % 100

def get_clicks_at_zero(raw_dial: int, prev_is_zero: bool) -> int:
    i = 1 if (raw_dial < 1 and not prev_is_zero) else 0
    return int(abs(raw_dial) / 100) + i
    
```

```python
# PART 1
dial = 50
result = 0

for cmd in data:
    dial = convert_rotation(rotate_dial(cmd, dial))
    if dial == 0:
        result += 1

result
```

```python
# PART 2
dial = 50
result = 0

for cmd in data:
    raw_dial = rotate_dial(cmd, dial)
    result += get_clicks_at_zero(raw_dial, dial == 0)
    dial = convert_rotation(raw_dial)

result
```
